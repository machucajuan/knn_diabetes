```python
# We start by importing the required libraries
import pandas as pd 
import numpy as np 
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
import seaborn as sn
```

```python
# KNN algorithm is based on the Euclidean Distance, which measures the lenght between two or more points
from IPython.display import Image
print("**Euclidean Distance Formula**")
Image(filename="euclidean distance.png", width= 650, height=250)
```

Why choosing the KNN algorithm? 

According to a PubMed survey, KNN is one most used machine learning models in the healthcare. Also, in a study of Krishnamoorthy et al., (2021) on diabetes prediction with multiple ML models, the KNN algorithm attained the highest accuracy score. 

```python
# Next, we import and read our data
df = pd.read_csv('diabetes.csv')
df.info(verbose=True)
print(len(df))
df.head()
```

The source of the data is the National Institute of Diabetes and Digestive and Kidney Diseases (USA). Given specific diagonostic measurements, our task is to predict whether or not the patient will have diabetes. For this purpose, we will use the KNN algorithm.

```python
# We start data prepping by replacing zero values
no_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'Insulin' ]

for column in no_zero:
    df[column] = df[column].replace(0, np.NaN)
    mean = int(df[column].mean(skipna=True))
    df[column] = df[column].replace(np.NaN, mean)

df['Insulin'].sort_values(ascending=False)
```

The predictor variables are number of pregnancies the patient has had, their BMI, insulin level, age, etc. Our target variable is the Outcome column, which classifies whether the patient has diabetes (1) or not (0). 

Attributes like "Glucose", "Blood Pressure", etc., with zero values will affect our model, thus, they are replaced with the mean

```python
# Let's conduct EDA to describe and summarize the data
diabetes_graphs = ['Pregnancies','Glucose','Insulin','BMI','DiabetesPedigreeFunction', 'Age','Outcome']
sn.set()
print(sn.pairplot(df[diabetes_graphs],hue='Outcome', diag_kind='kde'))
```

After our EDA and literature review, we conclude that all the attributes are relevant for our model as they contain essencial clinical records (Krishnamoorthy et al., 2021)

```python
#  Next, we split our data as a previous step to initalize our model
x = df.iloc[:, 0:8]
y = df.iloc[:,8]
x_train, x_test, y_train, y_test  = train_test_split(x,y, random_state=0, test_size=0.2)
```

```python
# We scale the values to fit them into our model
scale_x = StandardScaler()
x_train = scale_x.fit_transform(x_train)
x_test =  scale_x.transform(x_test)
```

```python
# Now it is time to choose the value of k (the number of nearest neighbors). A commonly used method is to use the square root of the number of observations in our data.
import math 
print(math.sqrt(len(y_train)))
print(math.sqrt(len(y_test))) # as 12 is even, we add -1 to make it odd, thus K = 12 - 1 = 11
```

A commonly used method to define the value of K is to use the square root of the number of observations in our data.

In this case, the result for our test data is 12. As even numbers are not recommended for the knn majority voting rule, we substract 1 to so that k is an odd number, thus: 
k = 12 - 1 = 11

```python
# We define the parameters and initialize our data: Init K-NN
classifier = KNeighborsClassifier(n_neighbors=11, p=2,metric='euclidean')
```

```python
# We continue by fitting the model
classifier.fit(x_train,y_train)
```

```python
# We run the model and we get the predicted results as the output
y_pred = classifier.predict(x_test)
y_pred
```

```python
# We evaluate the model
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(f1_score(y_test, y_pred))
```

To assess our model performance, we employ the confusion matrix. In this case, the rows represent the Actual Class, while the columns are the Predicted Class
Further, to evaluate the model accuracy, we employ the F1 score and the accuracy. As per the literature 

"The F1 score can be interpreted as a harmonic mean of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0."

F1 = 2 * (precision * recall) / (precision + recall)

```python
# Finally, we calculate the accuracy
print(accuracy_score(y_test,y_pred))
```

In this project, we solved a classification problem using the KNN to determine diabetes diagnose in a clinical records dataset. In this case, our model performed relatively well, with an accuracy of almost 82% and a F1 score of close to 70%. 

However, further model optimization and tuning is needed to obtain better precision and recall, thus improving overall performance.

Trying these models with huge datasets with minimal or no misplaced characteristic values will disclose added insight and healthier predictability (Krishnamoorthy et al., 2021).
